#Dimensionality Reduction 

```python
%matplotlib inline
import Sonar.SonarData as sd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.externals import joblib
from itertools import cycle
lines = ["-","--","-.",]
linecycle = cycle(lines)

fname = os.getenv('SONARDATAPATH') + '/lofar_data_1024fft_3dec.jbl'
nPts = 400
```

```python
# Load and filter data
data = joblib.load(fname)
for shipData in data.values():
    for runData in shipData.values():
        runData['Signal'] = runData['Signal'][:nPts]
        runData['Freqs'] = runData['Freqs'][:nPts]
```

##PCA

### Individually per Ship 

```python
from sklearn import decomposition
from sklearn import preprocessing
# Do it per run and considering all runs
plt.figure(figsize=(16,16))
for iship, ship in enumerate(data.keys()):
    runs = data[ship].keys()
    plt.subplot(2,2,iship+1)
    for irun, run in enumerate(runs):
        X = data[ship][run]['Signal']
        pca = decomposition.PCA().fit(X)
        plt.plot(np.cumsum(pca.explained_variance_ratio_),next(linecycle), lw=2, label='Run %i'%run)
    # Combine all
    Xall = np.concatenate([x['Signal'] for x in data[ship].values()], axis=1)
    pca = decomposition.PCA().fit(Xall)
    plt.plot(np.cumsum(pca.explained_variance_ratio_),next(linecycle), lw=2, label='All runs')
    plt.xlabel('# of Components')
    plt.ylabel('Cumulative Variance Rate')
    plt.title('Charge Curve for ' + ship)
    plt.grid()
    plt.ylim((0,1))
    plt.plot(plt.xlim(),[0.9,0.9],'--k',lw=3, label='0.9 Thr.')
    plt.legend(loc='best')
```

###All Ships Together 

```python
plt.figure(figsize=(6,6))
for ship in data.keys():
    Xall = np.concatenate([x['Signal'] for x in data[ship].values()], axis=1)
    pca = decomposition.PCA().fit(Xall)
    plt.plot(np.cumsum(pca.explained_variance_ratio_),next(linecycle), lw=2, label=ship)
X = np.concatenate([y['Signal']for x in data.values() for y in x.values()], axis=1)
pca = decomposition.PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_),next(linecycle), lw=2, label='All Ships')
plt.xlabel('# of Components')
plt.ylabel('Cumulative Variance Rate')
plt.title('Charge Curve')
plt.grid()
plt.ylim((0,1))
plt.plot(plt.xlim(),[0.9,0.9],'--k',lw=3, label='0.9 Thr.')
plt.legend(loc='best')
```
