# Compare PCD with Deflation and Cooperative models

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas
from sklearn.externals import joblib
from sklearn import metrics
from itertools import cycle, repeat
# Network import: should be in the path if framework configured correctly
import copy
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.stats import gaussian_kde
import scipy as sc
import itertools

np.set_printoptions(precision=2, suppress=True)

# Standard styles for each class
dashes = [[],[10,10],[10,4,2,4],[10,5,100,5]] 
colors = ['b','r','g','y']
markers= ['s','o','d','^']

import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['legend.handlelength'] = 3
mpl.rcParams['legend.borderpad'] = 0.3
mpl.rcParams['legend.numpoints'] = 1


classes = np.array(['ClasseA','ClasseB','ClasseC','ClasseD'])
noveltyclasses = np.array(['ClasseA','ClasseB','ClasseC','ClasseD'])
#noveltyclasses = np.array(['ClasseA'])
sonarhome = os.getenv('SONARHOME')
sonarnov = os.getenv('SONARNOVELTY')
nfft = 1024
nPts = 400
```

## Load Data

```python
data = {}
target = {}
data_nov = {}
cvPar = {}
nov_det_classes = {}
for nov in noveltyclasses:
    fdata = sonarhome + '/data/novelty_' + nov + '_' + str(nfft) + 'nfft.jbl'
    obj = joblib.load(fdata)
    data[nov] = obj['data']
    data_nov[nov] = obj['data_nov']
    target[nov] = obj['target']
    cvPar[nov] = obj['cvPar']
    nov_det_classes[nov] = obj['classes']
nPts = data[nov].shape[1]
```

## Load PCDs

```python
pcdnet = {}
for nov in noveltyclasses:
    pcdnet[nov] = {}
    # Deflation (or independent)
    filepcd = sonarnov + '/PCD/Deflation/pcddef_' + nov + '_' + str(nfft) + 'nfft.jbl'
    pcdnet[nov]['def'] = joblib.load(filepcd)
    # Cooperative (or constructive)
    filepcd = sonarnov + '/PCD/Constructive/pcdcons_' + nov + '_' + str(nfft) + 'nfft.jbl'
    pcdnet[nov]['coo'] = joblib.load(filepcd)
```

## PCD Efficiency - Deflation

```python
NCV = pcdnet.values()[0]['def']['cvPar']['CVNSel'] # it should be independent of which novelty
def_NMaxPCD = pcdnet.values()[0]['def']['pcdPar']['MaxPCD']
def_nov_avg_sps = np.zeros((len(pcdnet), def_NMaxPCD)) * np.nan
def_nov_std_sps = np.zeros((len(pcdnet), def_NMaxPCD)) * np.nan
def_nov_avg_pds = np.zeros((len(pcdnet), def_NMaxPCD)) * np.nan
def_nov_std_pds = np.zeros((len(pcdnet), def_NMaxPCD)) * np.nan
sps  = np.zeros((NCV, def_NMaxPCD)) * np.nan
pds  = np.zeros((NCV, def_NMaxPCD)) * np.nan
def_NumberOfPCDs = np.arange(1, def_NMaxPCD+1)
def_non_nov_pds = {} # dictionary per novelty. Efficiency for non-novelties 
lab = []
for inov, nov in enumerate(noveltyclasses):
    lab.append(nov)
    def_non_nov_pds[nov] = {}
    for nonnov in pcdnet[nov]['def']['Classes']:
        def_non_nov_pds[nov][nonnov] = np.zeros((NCV, def_NMaxPCD)) * np.nan
    for icv in range(NCV):
        pycv = pcdnet[nov]['def']['Classifier'].results[icv]
        if pycv is None:
            continue
        for ipcd in range(def_NMaxPCD):
            trninfo = pycv[ipcd+1].nnets.values()[0].trn_info
            ibst = trninfo.best_epoch if trninfo.best_epoch != 0 else -1
            for i, nonnov in enumerate(pcdnet[nov]['def']['Classes']):
                # indexes should be in the same order as novelty[nov]['Classes']
                def_non_nov_pds[nov][nonnov][icv,ipcd] = trninfo.metrics['eff_val_c'+str(i)][ibst]
            sps[icv,ipcd] = trninfo.metrics['sp_val'][ibst]
            pds[icv,ipcd] = np.mean([trninfo.metrics['eff_val_c0'][ibst],
                                     trninfo.metrics['eff_val_c1'][ibst],
                                     trninfo.metrics['eff_val_c2'][ibst]])
    def_nov_avg_sps[inov] = np.mean(sps, axis=0)
    def_nov_std_sps[inov] = np.std(sps, axis=0)
    def_nov_avg_pds[inov] = np.mean(pds, axis=0)
    def_nov_std_pds[inov] = np.std(pds, axis=0)
```

```python
plt.figure(figsize=(15,5), facecolor='w')
# PLOTS
# SP
plt.subplot(1,2,1, alpha=0)
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0][0] # the class index, not the novelty index
    plt.errorbar(def_NumberOfPCDs, def_nov_avg_sps[inov], def_nov_std_sps[inov],marker='o',
                 color=colors[icls], dashes=dashes[icls])
plt.grid(True) 
plt.ylim([0.4, 1])
plt.title('PCD Considering Each Novelty')
plt.xlabel('Number of PCD')
plt.ylabel('SP Index')
ax2 = plt.axes([.22, .18, .25, .53], axisbg ='w')
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0][0] # the class index, not the novelty index
    ax2.errorbar(def_NumberOfPCDs[2:], def_nov_avg_sps[inov][2:], def_nov_std_sps[inov][2:],marker='o',
                 color=colors[icls], dashes=dashes[icls])
ax2.grid(True)
plt.ylim([0.82,0.95])
plt.legend(lab, loc= 'lower right', ncol=2).get_frame().set_facecolor('1')
# PD
plt.subplot(1,2,2, alpha=0)
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0][0] # the class index, not the novelty index
    plt.errorbar(def_NumberOfPCDs, def_nov_avg_pds[inov], def_nov_std_pds[inov],marker='o',
                 color=colors[icls], dashes=dashes[icls])
plt.grid(True) 
plt.ylim([0.4, 1])
plt.title('PCD Considering Each Novelty')
plt.xlabel('Number of PCD')
plt.ylabel('Average Detection')
ax2 = plt.axes([.643, .18, .25, .5], axisbg ='w')
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0][0] # the class index, not the novelty index
    ax2.errorbar(def_NumberOfPCDs[2:],def_nov_avg_pds[inov][2:], def_nov_std_pds[inov][2:],marker='o',
                 color=colors[icls], dashes=dashes[icls])
ax2.grid(True)
plt.legend(lab, loc= 'lower right', ncol=2,).get_frame().set_facecolor('1')
plt.ylim([0.82,0.95]);
```

## PCD Efficiency - Cooperative

```python
NCV = pcdnet.values()[0]['coo']['cvPar']['CVNSel'] # it should be independent of which novelty
coo_NMaxPCD = pcdnet.values()[0]['coo']['pcdPar']['MaxPCD']
coo_nov_avg_sps = np.zeros((len(pcdnet), coo_NMaxPCD)) * np.nan
coo_nov_std_sps = np.zeros((len(pcdnet), coo_NMaxPCD)) * np.nan
coo_nov_avg_pds = np.zeros((len(pcdnet), coo_NMaxPCD)) * np.nan
coo_nov_std_pds = np.zeros((len(pcdnet), coo_NMaxPCD)) * np.nan
sps  = np.zeros((NCV, coo_NMaxPCD)) * np.nan
pds  = np.zeros((NCV, coo_NMaxPCD)) * np.nan
coo_NumberOfPCDs = np.arange(1, coo_NMaxPCD+1)
coo_non_nov_pds = {} # dictionary per novelty. Efficiency for non-novelties 
lab = []
for inov, nov in enumerate(noveltyclasses):
    lab.append(nov)
    coo_non_nov_pds[nov] = {}
    for nonnov in pcdnet[nov]['coo']['Classes']:
        coo_non_nov_pds[nov][nonnov] = np.zeros((NCV, coo_NMaxPCD)) * np.nan
    for icv in range(NCV):
        pycv = pcdnet[nov]['coo']['PCDModel'].results[icv]
        if pycv is None:
            continue
        for ipcd in range(coo_NMaxPCD):
            trninfo = pycv.PCDNets[ipcd].trn_info
            ibst = trninfo.best_epoch if trninfo.best_epoch != 0 else -1
            for i, nonnov in enumerate(pcdnet[nov]['coo']['Classes']):
                # indexes should be in the same order as novelty[nov]['Classes']
                coo_non_nov_pds[nov][nonnov][icv,ipcd] = trninfo.metrics['eff_val_c'+str(i)][ibst]
            sps[icv,ipcd] = trninfo.metrics['sp_val'][ibst]
            pds[icv,ipcd] = np.mean([trninfo.metrics['eff_val_c0'][ibst],
                                     trninfo.metrics['eff_val_c1'][ibst],
                                     trninfo.metrics['eff_val_c2'][ibst]])
    coo_nov_avg_sps[inov] = np.nanmean(sps, axis=0)
    coo_nov_std_sps[inov] = np.nanstd(sps, axis=0)
    coo_nov_avg_pds[inov] = np.nanmean(pds, axis=0)
    coo_nov_std_pds[inov] = np.nanstd(pds, axis=0)
```

## Performance Analysis

```python
npcd = 30
```

### Network Simulation

```python
Y = {}
for novcls in noveltyclasses:
    Y[novcls] = {}
    NCV = cvPar[novcls]['CVNSel']
    # Over CV
    for icv in range(NCV):
        Y[novcls][icv] = {}
        # For Deflation
        W = pcdnet[novcls]['def']['PCDModel'].results[icv].PCDNets[npcd-1].W[0]
        nnet = pcdnet[novcls]['def']['Classifier'].results[icv][npcd].nnets[npcd]
        Y[novcls][icv]['DefOutput'] = nnet.feedforward(W.dot(data[novcls].transpose()).transpose())
        Y[novcls][icv]['DefOutNov'] = nnet.feedforward(W.dot(data_nov[novcls].transpose()).transpose())
        # For Cooperative
        nnet = pcdnet[novcls]['coo']['PCDModel'].results[icv].PCDNets[npcd-1]
        Y[novcls][icv]['CooOutput'] = nnet.feedforward(data[novcls])
        Y[novcls][icv]['CooOutNov'] = nnet.feedforward(data_nov[novcls])
        # Test index
        Y[novcls][icv]['ITest'] = nnet.trn_info.ival
```

## Noise Sensitivity

```python
plt.figure(figsize=(15,9), facecolor='w')
noise_levels = np.linspace(0.1, 0.5, 2)
thrs = np.arange(-1,1.25,0.01)
for inov, novcls in enumerate(noveltyclasses):
    NCV = cvPar[novcls]['CVNSel']
    # Over different noise levels
    trg_def = np.zeros((len(noise_levels), len(thrs)))  # total efficiency (known-classes)    
    trg_coo = np.zeros((len(noise_levels), len(thrs)))  # total efficiency (known-classes)    
    for inoise, noise_sig in enumerate(noise_levels):
        # Over CV
        nsamples = data[novcls].shape[1]
        eff_def = np.zeros((NCV, len(thrs)))
        eff_coo = np.zeros((NCV, len(thrs)))
        noise_data = data[novcls] + np.random.normal(0, noise_sig, data[novcls].shape)
        for icv in range(NCV):
            # For Deflation
            W = pcdnet[novcls]['def']['PCDModel'].results[icv].PCDNets[npcd-1].W[0]
            nnet = pcdnet[novcls]['def']['Classifier'].results[icv][npcd].nnets[npcd]
            itst = nnet.trn_info.ival
            Ynoise = nnet.feedforward(W.dot(noise_data[itst].transpose()).transpose())
            eff_def[icv] = [np.sum((np.max(Ynoise, axis=1)>=thr))/float(itst.shape[0]) for thr in thrs]
            # For Cooperative
            itst = nnet.trn_info.ival
            nnet = pcdnet[novcls]['coo']['PCDModel'].results[icv].PCDNets[npcd-1]
            Ynoise = nnet.feedforward(noise_data[itst])
            eff_coo[icv] = [np.sum((np.max(Ynoise, axis=1)>=thr))/float(itst.shape[0]) for thr in thrs]
        # Average Trigger per Cross-Validation
        trg_def[inoise] = np.mean(eff_def, axis = 0)
        trg_coo[inoise] = np.mean(eff_coo, axis = 0)
    plt.subplot(2,2,inov+1)
    # Deflation
    icls = np.nonzero(classes == novcls)[0][0]
    for effs in trg_def: h1 = plt.plot(thrs, effs, '--sk', markevery=10)
    # System trigger (detection rate)
    effs = np.zeros((NCV, len(thrs)))*np.nan  # total efficiency (known-classes)    
    for icv in range(NCV):
        itst = Y[novcls][icv]['ITest']
        Yout = Y[novcls][icv]['DefOutput'][itst]
        effs[icv] = [np.sum((np.max(Yout, axis=1)>=thr)) / float(itst.shape[0]) for thr in thrs]
    h2 = plt.errorbar(thrs, np.mean(effs, axis=0), np.std(effs, axis=0), fmt='-or',
                      errorevery=10, markevery=10)
    # Cooperative
    icls = np.nonzero(classes == novcls)[0][0]
    for effs in trg_coo: h3 = plt.plot(thrs, effs, '--db', markevery=10)
    # System trigger (detection rate)
    effs = np.zeros((NCV, len(thrs)))*np.nan  # total efficiency (known-classes)    
    for icv in range(NCV):
        itst = Y[novcls][icv]['ITest']
        Yout = Y[novcls][icv]['CooOutput'][itst]
        effs[icv] = [np.sum((np.max(Yout, axis=1)>=thr)) / float(itst.shape[0]) for thr in thrs]
    h4 = plt.errorbar(thrs, np.mean(effs, axis=0), np.std(effs, axis=0), fmt='-^g',
                      errorevery=10, markevery=10)
    
    plt.axis([0, 1, 0.5, 1])
    plt.grid(True)
    plt.xlabel('Decision Threshold')
    plt.ylabel('Detection Efficiency')
    plt.title('Novelty: ' + novcls)
    plt.legend([h1[0],h2[0],h3[0],h4[0]],['Def. Noise', 'Def. Trigger', 'Coo. Noise', 'Coo. Trigger'],
               loc= 'lower left', ncol=2,).get_frame().set_facecolor('1')
```

## Final PCD 

From the cross validation analysis, we can estimate the statistical effect on the PCD performance, but we still need to find a deterministic basis for the PCD transformation. Only one network out of the N cross validation tests is elected the winner and used in later analysis. This selection depends on the purpose of the PCD.

Although the PCD aims for detection, the idea is to find the best discrimination directions that characterizes the known classes. The novelty itself is later analyzed considering a neural classifier over the data projection onto the PCD. For that reason, the final PCD will be selected considering the network that best characterizes the detection of the entire dataset of the known classes.

```python
finalPCD = {'def':{}, 'coo':{}}
finalNet = {'def':{}, 'coo':{}}
finalSim = {'def':{}, 'coo':{}}
for inov, novcls in enumerate(noveltyclasses):
    finalNet['def'][novcls] = None
    finalNet['coo'][novcls] = None
    finalPCD['def'][novcls] = None
    finalPCD['coo'][novcls] = None
    finalSim['def'][novcls] = None
    finalSim['coo'][novcls] = None
    sps = np.zeros((2, len(Y[novcls])))*np.nan
    tpr = np.zeros((2, len(Y[novcls])))*np.nan
    mse = np.zeros((2, len(Y[novcls])))*np.nan
    for icv in range(len(Y[novcls])):
        T = target[novcls]
        #itst = Y[novcls][icv]['ITest']
        itst = np.array([True] * T.shape[0])
        def_effs = np.zeros(len(pcdnet[novcls]['def']['Classes']))
        coo_effs = np.zeros(len(pcdnet[novcls]['coo']['Classes']))
        def_err = 0
        coo_err = 0
        for iout in range(len(pcdnet[novcls]['def']['Classes'])):
            # Deflation
            TP = np.sum((np.argmax(Y[novcls][icv]['DefOutput'][itst], axis=1) == iout) & (T[itst, iout] == 1))
            def_effs[iout] = TP / float(np.sum((T[itst, iout] == 1)))
            def_err = np.sum(np.power(Y[novcls][icv]['DefOutput'][itst, iout] - T[itst, iout], 2)) + def_err
            # Cooperative
            TP = np.sum((np.argmax(Y[novcls][icv]['CooOutput'][itst], axis=1) == iout) & (T[itst, iout] == 1))
            coo_effs[iout] = TP / float(np.sum((T[itst, iout] == 1)))
            coo_err = np.sum(np.power(Y[novcls][icv]['CooOutput'][itst, iout] - T[itst, iout], 2)) + coo_err
        mse[0, icv] = def_err / itst.shape[0]
        mse[1, icv] = coo_err / itst.shape[0]
        sps[0, icv] = np.sqrt(np.power(np.prod(def_effs), 1./len(def_effs)) * np.mean(def_effs))
        sps[1, icv] = np.sqrt(np.power(np.prod(coo_effs), 1./len(coo_effs)) * np.mean(coo_effs))
        tpr[0, icv] = np.mean(def_effs)
        tpr[1, icv] = np.mean(coo_effs)
    iwinner_def = np.nanargmax(sps[0])
    iwinner_coo = np.nanargmax(sps[1])
    finalPCD['def'][novcls] = pcdnet[novcls]['def']['PCDModel'].results[iwinner_def].PCDNets[npcd-1]
    finalNet['def'][novcls] = pcdnet[novcls]['def']['Classifier'].results[iwinner_def][npcd].nnets[npcd]
    finalSim['def'][novcls] = Y[novcls][iwinner_def]
    print novcls,':',
    print '\tDef, SP = %.3f, PD = %.3f, MSE = %.3f for #cv %i'%(sps[0,iwinner_def]*100,
                                                           tpr[0,iwinner_def]*100,
                                                           mse[0,iwinner_def],
                                                          iwinner_def)
    finalPCD['coo'][novcls] = pcdnet[novcls]['coo']['PCDModel'].results[iwinner_coo].PCDNets[npcd-1]
    finalNet['coo'][novcls] = pcdnet[novcls]['coo']['PCDModel'].results[iwinner_coo].PCDNets[npcd-1]
    finalSim['coo'][novcls] = Y[novcls][iwinner_coo]
    print '\t\tCoo, SP = %.3f, PD = %.3f, MSE = %.3f for #cv %i'%(sps[1,iwinner_coo]*100,
                                                           tpr[1,iwinner_coo]*100,
                                                           mse[1,iwinner_coo],
                                                          iwinner_coo)    
```

### PCD Directions

#### Angles

```python
Wboth = np.vstack((Wdef, Wcoo))
Cang = np.zeros((Wboth.shape[0], Wboth.shape[0])) * np.nan
for i in range(0, Cang.shape[0]-1):
    for j in range(i+1, Cang.shape[1]):
        dot = Wboth[i,:].dot(Wboth[j,:])
        norm_i = np.linalg.norm(Wboth[i,:],2)
        norm_j = np.linalg.norm(Wboth[j,:],2)
        Cang[i,j] = np.degrees(np.arccos(dot/norm_i/norm_j))
        Cang[i,j] = 0 if np.isnan(Cang[i,j]) else Cang[i,j]
Cang[(Cang > 90)&(Cang <= 180)] = Cang[(Cang > 90)&(Cang < 180)] - 180
Cang[(Cang > 180)&(Cang <= 270)]= Cang[(Cang > 180)&(Cang <= 270)] - 180
Cang = np.abs(Cang)

plt.figure(figsize=(20,20))
color_helper = 'k'
plt.plot([Wdef.shape[0]-1+0.5, Wdef.shape[0]-1+0.5], [-1, Wboth.shape[0]], '-'+color_helper, lw=4)
plt.plot([-1, Wboth.shape[0]], [Wdef.shape[0]-1+0.5, Wdef.shape[0]-1+0.5], '-'+color_helper, lw=4)
plt.plot([Wdef.shape[0]-1+0.5, Wboth.shape[0]-1 + 0.5],[-0.5, Wdef.shape[0]-1+0.5], '-'+color_helper, lw=4)
ms = plt.matshow(Cang, cmap=plt.cm.jet, vmin=0, vmax=90, fignum=0)

if False:
    for i in range(Cang.shape[0]-1):
        for j in range(i+1,Cang.shape[1]):
            color = 'k' if Cang[i,j] < 50 else 'w'
            plt.text(j,i,'%i'%(Cang[i,j]), ha='center', va='center',size=14, color=color, weight='bold')
plt.xlabel('#PCD')
plt.ylabel('#PCD')
plt.title('Angles (Deflation, Cooperative) - ' + workcls)
plt.gca().set_xticks(np.arange(Wboth.shape[0]))
plt.gca().set_xticklabels(np.hstack((np.arange(Wdef.shape[0]), np.arange(Wcoo.shape[0]))))
plt.gca().set_yticks(np.arange(Wboth.shape[0]))
plt.gca().set_yticklabels(np.hstack((np.arange(Wdef.shape[0]), np.arange(Wcoo.shape[0]))))
plt.tick_params(labelbottom=True, labelright=True)

divider = make_axes_locatable(plt.gca())
cax = divider.append_axes("right", "3%", pad="2%")
plt.colorbar(ms, cax=cax);
```

#### Mutual Information

```python
import sklearn.metrics as metrics

Xpcd_both = pandas.concat((Xpcd_def, Xpcd_coo), axis=1)
Cmutual = np.zeros((Xpcd_both.shape[1], Xpcd_both.shape[1])) * np.nan
for i in range(0, Xpcd_both.shape[1]-1):
    for j in range(i+1, Xpcd_both.shape[1]):
        xbins = np.linspace(Xpcd_both.iloc[:,i].min(), Xpcd_both.iloc[:,i].max(), 50)
        ybins = np.linspace(Xpcd_both.iloc[:,j].min(), Xpcd_both.iloc[:,j].max(), 50)
        dsx, dsy = np.meshgrid(xbins, ybins)
        ds = gaussian_kde(Xpcd_both.iloc[np.random.randint(0,Xpcd_both.shape[0], 1000), [i,j]].transpose())
        pos = np.vstack([dsx.ravel(),dsy.ravel()])
        dsz = np.reshape(ds(pos).T, (xbins.shape[0], ybins.shape[0]))
        Cmutual[i,j] = metrics.mutual_info_score(None, None, contingency=dsz)
# PLOT
plt.figure(figsize=(20,20))
color_helper = 'k'
plt.plot([Xpcd_def.shape[1]-1+0.5, Xpcd_def.shape[1]-1+0.5], [-1, Cmutual.shape[0]], '-'+color_helper, lw=4)
plt.plot([-1, Cmutual.shape[0]], [Xpcd_def.shape[1]-1+0.5, Xpcd_def.shape[1]-1+0.5], '-'+color_helper, lw=4)
plt.plot([Xpcd_def.shape[1]-1+0.5, Cmutual.shape[1]-1 + 0.5],[-0.5, Xpcd_def.shape[1]-1+0.5], '-'+color_helper, lw=4)
ms = plt.matshow(Cmutual, cmap=plt.cm.jet, vmin=0, vmax=np.nanmax(Cmutual), fignum=0)

if False:
    for i in range(Cmutual.shape[0]-1):
        for j in range(i+1,Cmutual.shape[1]):
            color = 'k' if Cmutual[i,j] < 0.5 else 'w'
            plt.text(j,i,'%.2f'%(Cmutual[i,j]), ha='center', va='center',size=10, color=color)
plt.xlabel('#PCD')
plt.ylabel('#PCD')
plt.title('Mutual Information (Deflation, Cooperative) - ' + workcls)
plt.gca().set_xticks(np.arange(Cmutual.shape[0]))
plt.gca().set_xticklabels(np.hstack((np.arange(Xpcd_def.shape[1]), np.arange(Xpcd_coo.shape[1]))))
plt.gca().set_yticks(np.arange(Cmutual.shape[0]))
plt.gca().set_yticklabels(np.hstack((np.arange(Xpcd_def.shape[1]), np.arange(Xpcd_coo.shape[1]))))
plt.tick_params(labelbottom=True, labelright=True)

divider = make_axes_locatable(plt.gca())
cax = divider.append_axes("right", "3%", pad="2%")
plt.colorbar(ms, cax=cax);
        
```
