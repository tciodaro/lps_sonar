# Train an ART Network for the Sonar Data

```python
%load_ext autoreload
%autoreload 3


%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas
from sklearn.externals import joblib
from sklearn import metrics
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.stats import gaussian_kde
import scipy as sc

import PyNN.ARTNet as pyart

np.set_printoptions(precision=2, suppress=True)

# Standard styles for each class
dashes = np.array([[],[10,10],[10,4,2,4],[10,5,100,5]] )
colors = np.array(['b','r','g','y'])
markers= np.array(['s','o','d','^'])

import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['legend.handlelength'] = 3
mpl.rcParams['legend.borderpad'] = 0.3
mpl.rcParams['legend.numpoints'] = 1


classes = np.array(['ClasseA','ClasseB','ClasseC','ClasseD'])
noveltyclasses = np.array(['ClasseA','ClasseB','ClasseC','ClasseD'])
#noveltyclasses = np.array(['ClasseA'])
sonarhome = os.getenv('SONARHOME')
sonarnov = os.getenv('SONARNOVELTY')
nfft = 1024
nPts = 400
```

# Load Data 

```python
data = {}
target = {}
data_nov = {}
cvPar = {}
for nov in noveltyclasses:
    fdata = sonarhome + '/data/novelty_' + nov + '_' + str(nfft) + 'nfft.jbl'
    obj = joblib.load(fdata)
    data[nov] = obj['data']
    data_nov[nov] = obj['data_nov']
    target[nov] = obj['target']
    cvPar[nov] = obj['cvPar']
```

# Load PCD

```python
pcdnet = {}
for nov in noveltyclasses:
    pcdnet[nov] = {}
    # Deflation (or independent)
    filepcd = sonarnov + '/PCD/Deflation/pcddef_' + nov + '_' + str(nfft) + 'nfft.jbl'
    pcdnet[nov]['def'] = joblib.load(filepcd)
    # Cooperative (or constructive)
    filepcd = sonarnov + '/PCD/Constructive/pcdcons_' + nov + '_' + str(nfft) + 'nfft.jbl'
    pcdnet[nov]['coo'] = joblib.load(filepcd)
 
```

# ART Pre-training

```python
workcls = 'ClasseA'
npcd = 6 # 0 == all
icv = 0
Wdef = pcdnet[workcls]['def']['PCDModel'].results[icv].PCDNets[npcd-1].W[0]
Wcoo = pcdnet[workcls]['coo']['PCDModel'].results[icv].PCDNets[npcd-1].W[0]
class_names = pcdnet[workcls]['def']['Classes']
T = target[workcls]
X = data[workcls]
Xnov = data_nov[workcls]
Xpcd_def = pandas.DataFrame(Wdef.dot(X.transpose()).transpose())
XnovPcd_def = pandas.DataFrame(Wdef.dot(Xnov.transpose()).transpose())
Xpcd_coo = pandas.DataFrame(Wcoo.dot(X.transpose()).transpose())
XnovPcd_coo = pandas.DataFrame(Wcoo.dot(Xnov.transpose()).transpose())
```

## Class Distribution Over PCDs

### PCD Cooperative 

```python
Nx = 3
Ny = int(np.floor(Xpcd_coo.shape[1]/float(Nx)))+1
plt.figure(figsize=(Nx*5,Ny*5), facecolor='w')
for ipcd in range(Xpcd_coo.shape[1]):
    plt.subplot(Ny, Nx, ipcd+1)
    for cls in range(len(pcdnet[workcls]['coo']['Classes'])):
        icls = np.nonzero(classes == pcdnet[workcls]['coo']['Classes'][cls])[0][0]
        T = target[workcls]
        idx = np.nonzero(T[:, cls] == 1)[0]
        Xpcd_coo.loc[idx, ipcd].plot(kind='kde', color = colors[icls],sharex=False,
                                     dashes=dashes[icls], label=pcdnet[workcls]['coo']['Classes'][cls])
    XnovPcd_coo.loc[:, ipcd].plot(kind='kde', color = 'k',sharex=False,label='Novelty')
    plt.title('Novelty ' + workcls +', PCD #'+str(ipcd+1))
    plt.legend(numpoints=1, loc='best').get_frame().set_facecolor('1')
    plt.grid(True)
```

# ART Training

```python
from sklearn import preprocessing 
############################################################################
## ART Parameters
trnPar = {
    'itrn': cvPar[workcls]['Indexes'][icv]['ITrn'],
    'ival': cvPar[workcls]['Indexes'][icv]['IVal'],
    'itst': cvPar[workcls]['Indexes'][icv]['ITst'],
    'trn_phase1'             : True,
    'trn_phase2'             : True,
    'trn_initial_radius'     : 0,     # Neuron initial radius
    'trn_max_radius_sample'  : 0.5 ,   # Number of samples to estimate radius
    'trn_radius_factor'      : 0.8 ,  # Factor applied to the initial radius
    'trn_sim_function'       : 'euclidean',
    'trn_eta'                : 0.1 ,  # Neuron update constant
    'trn_eta_decay'          : 0.01, # Regularization factor
    'trn_nshow'              : 0 ,    # Number of iterations to show
    'trn_tol'                : 1e-6 , # Min allowed tolerance to consider the stall
    'trn_max_stall'          : 10 ,   # Max allowed iterations on stall
    'trn_max_neurons_rate'   : 0.7 ,  # Percentage of neurons wrt data events
    'trn_max_it'             : 100 ,  # Max number of iterations
    'trn_max_no_new_neuron'  : 10 ,   # Max number of iterations without neuron creation
    'trn_phase2_max_eff'     : 0.05,
    'trn_data_norm' : None #preprocessing.StandardScaler()
}    
```
