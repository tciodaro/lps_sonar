```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
%matplotlib inline
colors = np.array(['r','k','y', 'b'])
import PyNN.ARTNet as PyART
import numpy as np


N = 1000
###############################################################################
## MAKE DATA
C1 = np.random.rand(N, 2)*2 +1
C2 = np.random.rand(N, 2)*4 -4
C3 = np.random.rand(N, 2)*2

#C1[:,0] = C1[:,0] * 10
#C2[:,0] = C2[:,0] * 10
#C3[:,0] = C3[:,0] * 10

X = np.concatenate((C1,C2,C3))
T = np.concatenate((np.zeros(N), np.ones(N), 2*np.ones(N)))

###############################################################################
## Training Indexes
indexes = np.arange(X.shape[0])
np.random.shuffle(indexes)
ntrn = 0.70
itrn = indexes[:int(X.shape[0]*ntrn)]
ival = indexes[int(X.shape[0]*(1-ntrn)):]
itst = ival
```

# ART - Train

```python
############################################################################
## ART Parameters
trnPar = {
    'itrn': itrn,
    'ival': ival,
    'itst': itst,
    'trn_phase1'             : True,
    'trn_phase2'             : True,
    'trn_phase3'             : True,
    'trn_initial_radius'     : 0,     # Neuron initial radius
    'trn_max_radius_sample'  : 0.5 ,   # Number of samples to estimate radius
    'trn_radius_factor'      : 1.0 ,  # Factor applied to the initial radius
    'trn_sim_function'       : 'euclidean',
    'trn_eta'                : 0.1 ,  # Neuron update constant
    'trn_eta_decay'          : 0.001, # Regularization factor
    'trn_nshow'              : 0 ,    # Number of iterations to show
    'trn_tol'                : 1e-6 , # Min allowed tolerance to consider the stall
    'trn_max_stall'          : 10 ,   # Max allowed iterations on stall
    'trn_data_norm'          : None , # Data normalizer
    'trn_max_neurons_rate'   : 0.7 ,  # Percentage of neurons wrt data events
    'trn_max_it'             : 100 ,  # Max number of iterations
    'trn_max_no_new_neuron'  : 10 ,   # Max number of iterations without neuron creation
    'trn_phase2_max_eff'     : 0.05,
    'trn_data_scaler' : None #preprocessing.StandardScaler()
}
```

```python
idx = T == 0
X[idx].shape
```

```python
nnet = PyART.ARTNet(trnPar)

if False:
    OneClassTarget = np.zeros((X.shape[0],1))
    nnet.train(X, OneClassTarget, trnPar)
    nnet.recolor(X, T)
else:
    nnet.train(X, T, trnPar)
    nnet.recolor(X, T, trnPar)
    nnet.performance(X[trnPar['ival']], T[trnPar['ival']])
```

```python
###############################################################################
## PLOT DATA
plt.figure(figsize=(6,6))
plt.plot(C1[:,0], C1[:,1], 'x'+colors[0])
plt.plot(C2[:,0], C2[:,1], 'x'+colors[1])
plt.plot(C3[:,0], C3[:,1], 'x'+colors[2])
###############################################################################
## PLOT NEURONS AND RADIUS
for ineuron in range(nnet.neurons.shape[0]):
    cls = nnet.classes[ineuron]
    c = colors[int(cls)] if cls != -1 else 'b'
    plt_neuron = plt.Circle(nnet.neurons[ineuron],nnet.radius[ineuron],color=c,fill=False)
    plt.gca().add_artist(plt_neuron)
    
###############################################################################
## BUSIEST NEURON
busiest = np.argmax(nnet.neuron_hits)
#plt_neuron = plt.Circle(nnet.neurons[busiest],nnet.radius[busiest],color='b',fill=False)
#plt.gca().add_artist(plt_neuron)
plt.plot(nnet.neurons[busiest,0], nnet.neurons[busiest,1], 'sb')
###############################################################################
## LASIEST NEURON
lasiest = np.argmin(nnet.neuron_hits)
#plt_neuron = plt.Circle(nnet.neurons[lasiest],nnet.radius[lasiest],color='b',fill=False)
#plt.gca().add_artist(plt_neuron)
plt.plot(nnet.neurons[lasiest,0], nnet.neurons[lasiest,1], 'db')
plt.tight_layout()
```
