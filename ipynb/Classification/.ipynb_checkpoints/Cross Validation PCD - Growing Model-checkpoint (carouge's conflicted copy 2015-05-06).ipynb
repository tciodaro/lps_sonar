# Cross Validation Analysis for Principal Components of Discrimination

## PCD

The PCD extraction is done by adding neurons in the hidden layer as the SP index increases more than a certain value. This is implemented by a 3 layers feed forward neural network (input, hidden and output layers).

## Cross Validation

The cross validation is done by randomly dividing the data into a number of boxes and selecting random boxes as train, test and validation sets. The data is considered for each acquisition run, of every ship. That is, the boxes and selections are done for each acquisition run. Later, they are appended in order to train the classifier with data from every acquisition. This guarantees that most of the signal stochasticity is seen by the classifier.

## Training

For each number of PCD and for each cross validation selection, the model is trained several times, for each random initialization, in order to avoid local minima.

The data used in the cross validation are the time windows from all signals, independently of acquisition run and ship. It is likelty that train, validation and test will have signals from every run and every ship.

# Results

```python
%matplotlib inline
import Sonar.SonarData as sd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.externals import joblib
from sklearn import metrics
from itertools import cycle, repeat
# Network import: should be in the path if framework configured correctly
import neuralnet as nn
import PyNN.NeuralNet as PyNNet
import PyNN.TrnInfo as PyTrnInfo
import copy

# Standard styles for each class
dashes = [[],[10,10],[10,4,2,4],[10,5,100,5]] 
colors = ['b','r','g','y']

import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['legend.handlelength'] = 3
mpl.rcParams['legend.borderpad'] = 0.3
mpl.rcParams['legend.numpoints'] = 1


classes = np.array(['ClasseA','ClasseB','ClasseC','ClasseD'])
noveltyclasses = np.array(['ClasseA','ClasseB','ClasseC','ClasseD'])
sonarhome = os.getenv('SONARHOME')
#fname = sonarhome + '/data/lofar_data_1024fft_3dec.jbl'
fname = sonarhome + '/data/lofar_data_1024fft_3dec_fromMat.jbl'
nPts = 400
```

```python
# Load and filter data
globaldata = joblib.load(fname)
def novelty_data(clsnov, rawdata):
    # Load and filter data
    data = copy.deepcopy(rawdata)
    for shipData in data.values():
        for runData in shipData.values():
            runData['Signal'] = runData['Signal'][:nPts,:]
            runData['Freqs'] = runData['Freqs'][:nPts]
    X = np.concatenate([y['Signal']for cls, x in data.items() for y in x.values() if cls in clsnov], axis=1)
    Xnov = X.transpose() # Transpose X, so events are rows (T is already correct)
    # Pop classes not used
    for ship in clsnov:
        data.pop(ship)
    X = None
    T = None
    nClass = len(data.values())
    classes = np.sort(data.keys())
    for iship, ship in enumerate(classes):
        tot = np.sum([runData['Signal'].shape[1] for runData in data[ship].values()])
        if nClass == 2:
            aux = -np.ones(1)
            aux[0] = 1 if iship > 0 else -1
        else:
            aux = -np.ones(nClass)
            aux[iship] = 1

        T = np.tile(aux,(tot,1)) if T is None else np.concatenate((T, np.tile(aux,(tot,1))))
    X = np.concatenate([y['Signal']for x in data.values() for y in x.values()], axis=1)
    X = X.transpose() # Transpose X, so events are rows (T is already correct)
    return(X, T, Xnov)
```

## Novelties

```python
novelty = {}
for cls in noveltyclasses:
    loadcls = np.setdiff1d(classes, [cls])
    #filepcd = sonarhome + '/results/classification/novelty/pcd_cv_' + ''.join(loadcls) + '_SP_RP.jbl'
    #filepcd = sonarhome + '/results/classification/novelty/pcd_cv_' + ''.join(loadcls) + '_SP_RP_replicate.jbl'
    filepcd = sonarhome + '/results/classification/novelty/PCD_Repl/pcd_cv_' + ''.join(loadcls) + '_SP_RP_fromMat_replicate.jbl'
    novelty[cls] = joblib.load(filepcd)
```

## PCD Efficiency

```python
NMaxPCD = novelty.values()[0]['CV'][0]['PCDMax'] 
nov_avg_sps = np.zeros((len(novelty), NMaxPCD))
nov_std_sps = np.zeros((len(novelty), NMaxPCD))
nov_avg_pds = np.zeros((len(novelty), NMaxPCD))
nov_std_pds = np.zeros((len(novelty), NMaxPCD))
NumberOfPCDs = np.arange(1, NMaxPCD+1)
non_nov_pds = {} # dictionary per novelty. Efficiency for non-novelties 
lab = []
for inov, nov in enumerate(noveltyclasses):
    lab.append(nov)
    NCV = novelty[nov]['CVSel'] # unlikely that these parameters change
    sps  = np.zeros((NCV, NMaxPCD))
    pds  = np.zeros((NCV, NMaxPCD))
    non_nov_pds[nov] = {}
    for nonnov in novelty[nov]['Classes']:
        non_nov_pds[nov][nonnov] = np.zeros((NCV, NMaxPCD))
    for icv in range(NCV):
        results = novelty[nov]['CV'][icv]
        if len(results['PCDNets']) != NMaxPCD:
            continue
        for ipcd in range(NMaxPCD):
            trninfo = results['PCDNets'][ipcd].trn_info
            ibst = trninfo.best_epoch if trninfo.best_epoch != 0 else -1
            for i, nonnov in enumerate(novelty[nov]['Classes']):
                # indexes should be in the same order as novelty[nov]['Classes']
                non_nov_pds[nov][nonnov][icv,ipcd] = trninfo.metrics['eff_val_c'+str(i)][ibst]
            sps[icv,ipcd] = trninfo.metrics['sp_val'][ibst]
            pds[icv,ipcd] = np.mean([trninfo.metrics['eff_val_c0'][ibst],
                                     trninfo.metrics['eff_val_c1'][ibst],
                                     trninfo.metrics['eff_val_c2'][ibst]])
    sps[np.nonzero(sps == 0)] = np.nan
    pds[np.nonzero(sps == 0)] = np.nan
    nov_avg_sps[inov] = np.nanmean(sps, axis=0)
    nov_std_sps[inov] = np.nanstd(sps, axis=0)
    nov_avg_pds[inov] = np.nanmean(pds, axis=0)
    nov_std_pds[inov] = np.nanstd(pds, axis=0)
```

```python
plt.figure(figsize=(15,5))
# PLOTS
# SP
plt.subplot(1,2,1, alpha=0)
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0] # the class index, not the novelty index
    plt.errorbar(NumberOfPCDs, nov_avg_sps[inov], nov_std_sps[inov], color=colors[icls], dashes=dashes[icls])
plt.grid(True) 
plt.ylim([0, 1])
plt.title('PCD Considering Each Novelty')
plt.xlabel('Number of PCD')
plt.ylabel('SP Index')
ax2 = plt.axes([.2, .2, .26, .5], axisbg ='w')
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0] # the class index, not the novelty index
    ax2.errorbar(NumberOfPCDs[1:],nov_avg_sps[inov][1:], nov_std_sps[inov][1:],
                 color=colors[icls], dashes=dashes[icls])
ax2.grid(True)
plt.legend(lab, loc= 'lower right', ncol=2).get_frame().set_facecolor('1')
# PD
plt.subplot(1,2,2, alpha=0)
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0] # the class index, not the novelty index
    plt.errorbar(NumberOfPCDs, nov_avg_pds[inov], nov_std_pds[inov], color=colors[icls], dashes=dashes[icls])
plt.grid(True) 
plt.ylim([0, 1])
plt.title('PCD Considering Each Novelty')
plt.xlabel('Number of PCD')
plt.ylabel('Average Detection')
ax2 = plt.axes([.62, .2, .27, .5], axisbg ='w')
for inov, nov in enumerate(noveltyclasses):
    icls = np.nonzero(classes == nov)[0] # the class index, not the novelty index
    ax2.errorbar(NumberOfPCDs[1:],nov_avg_pds[inov][1:], nov_std_pds[inov][1:],
                 color=colors[icls], dashes=dashes[icls])
ax2.grid(True)
plt.legend(lab, loc= 'lower right', ncol=2,).get_frame().set_facecolor('1')
```

## PCD Class Detection Per Novelty 

```python
plt.figure(figsize=(15,10))
for inov, nov in enumerate(novelty.keys()):
    plt.subplot(2,2,inov+1)
    # Loop over detected classes
    for nonnov in non_nov_pds[nov]:
        icls = np.nonzero(classes == nonnov)[0] # the class index, not the novelty index
        avg_pds = np.mean(non_nov_pds[nov][nonnov], axis=0)
        std_pds = np.std(non_nov_pds[nov][nonnov], axis=0)        
        plt.errorbar(NumberOfPCDs[1:],avg_pds[1:], std_pds[1:],color=colors[icls], dashes=dashes[icls], label=nonnov)
    plt.ylim([0.7,1])
    plt.grid()
    plt.xlabel('Number of PCD')
    plt.ylabel('Detection Efficiency')
    plt.title('Novelty: ' + nov)
    plt.legend(loc='lower right',ncol=3).get_frame().set_facecolor('1')
```

## Performance Analysis

```python
npcd = 4
```

### Network Simulation

```python
Y = {}
for novcls in noveltyclasses:
    Y[novcls] = {}
    NCV = novelty[novcls]['CVSel']
    # Over CV
    for icv in range(NCV):
        Y[novcls][icv] = {}
        results = novelty[novcls]['CV'][icv] 
        nnet = results['PCDNets'][npcd-1]
        Y[novcls][icv]['Labels'] = novelty[novcls]['Classes']
        X, T, Xnov = novelty_data([novcls], globaldata)
        itst = nnet.trn_info.itst
        Y[novcls][icv]['Output'] = nnet.feedforward(X[itst])
        Y[novcls][icv]['Novelty'] = nnet.feedforward(Xnov)
        Y[novcls][icv]['Target'] = T[itst]
```

### Performance Over Threshold

```python
plt.figure(figsize=(15,8))
thrs = np.linspace(-1, 1, 100)
# Loop over novelties
for inov, novcls in enumerate(noveltyclasses):
    plt.subplot(2,2,inov+1)
    NCV = novelty[novcls]['CVSel']
    # PLot Known classes
    for iout in range(len(novelty[novcls]['Classes'])):
        # Loop over cross-validation
        tpr = np.zeros((NCV, len(thrs)))  # true positive rate
        effnov = np.zeros((NCV, len(thrs)))  # novely efficiency
        for icv in range(NCV):
            Yout = Y[novcls][icv]['Output']
            T = Y[novcls][icv]['Target']
            # Loop over thresholds
            for ithr, thr in enumerate(thrs):
                # True Positive: greater than the threshold and the other neurons.
                TP = np.sum((Yout[:, iout]>=thr) & (np.argmax(Yout, axis=1) == iout) & (T[:, iout] == 1))
                tpr[icv,ithr] = TP / float(np.sum((T[:, iout] == 1)))
        # Plot
        icls = np.nonzero(classes == novelty[novcls]['Classes'][iout])[0]
        plt.errorbar(thrs, np.mean(tpr, axis=0), np.std(tpr, axis=0), dashes=dashes[icls],color=colors[icls],
                     errorevery=10, label=novelty[novcls]['Classes'][iout])
    # Plot Novelty
    for icv in range(NCV):
        Ynov = Y[novcls][icv]['Novelty']
        effnov[icv] = np.array([np.sum(np.max(Ynov, axis=1) < thr) for thr in thrs], 'f')/Ynov.shape[0]
    plt.errorbar(thrs, np.mean(effnov, axis=0), np.std(effnov, axis=0),fmt='k-', errorevery=10, label='Novelty')
    # Plot aspects
    plt.gca().set_yticks(np.linspace(0,1,11))
    plt.gca().set_xticks(np.arange(-1,1.25,0.25))
    plt.xlim([-1,1])
    plt.ylim([0,1])
    plt.grid()
    plt.xlabel('Decision Threshold')
    plt.ylabel('Detection Efficiency')
    plt.legend(loc='center left').get_frame().set_facecolor('1')
```
