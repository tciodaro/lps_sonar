# Intrinsic Dimensions in the Oil Data (Crude)

Several properties describe the oil samples for a given distillation cut. Not necessarily this properties are independent,
or even necessary to fully characterize a the oil sample. Through the Isomap algorithm it is possible to use a low dimensional
mapping considering geodesic distances to fully characterize the oil sample, possibly reducing the input space dimensionality.

```python
%matplotlib inline
from Blend.Isomap import *
import numpy as np
import scipy as sc
from sklearn import manifold
from sklearn import cluster
from sklearn import metrics
import itertools
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt
import pandas

# LOAD DATA
data = pandas.read_pickle('/Users/ciodaro/Projects/BlendBR/Data/crude_data.pck')
```

```python
def plot_silhouette(svalues, clusters):
    n_clusters = len(np.unique(clusters))
    y_lower = 10
    # The silhouette coefficient can range from -1, 1 but in this example all
    plt.gca().set_xlim([np.min(svalues), 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    plt.gca().set_ylim([0, len(svalues) + (n_clusters + 1) * 10])
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = svalues[clusters == i]
        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = plt.cm.spectral(float(i) / n_clusters)
        plt.gca().fill_betweenx(np.arange(y_lower, y_upper),
                         0, ith_cluster_silhouette_values,
                         facecolor=color, edgecolor=color, alpha=0.7)
        # Label the silhouette plots with their cluster numbers at the middle
        plt.gca().text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    plt.gca().set_title("The silhouette plot for the various clusters.")
    plt.gca().set_xlabel("The silhouette coefficient values")
    plt.gca().set_ylabel("Cluster label")
    plt.gca().set_yticks([])  # Clear the yaxis labels / ticks
    
```

```python
from scipy.stats import gaussian_kde
def kdeplot(data, covfac=0, hist = True):
    X = np.arange(np.min(data), np.max(data), (np.max(data)-np.min(data)) * 0.01)
    density = gaussian_kde(data)
    if covfac:
        density.covariance_factor = lambda : covfac
        density._compute_covariance()
    h = None
    if hist:
        N = int((np.max(data)-np.min(data))/(0.33*np.std(data)))
        h = plt.hist(data, bins=N, normed=True, facecolor='#777777', edgecolor='#777777', label='Data')
    hd = plt.plot(X,density(X), '-', lw=3, label='Estimation')
    plt.xlabel('Data Values')
    plt.ylabel('Normalized Count')
    plt.legend(loc='best')
    plt.title('KDE - Gaussian (Var.: %.3e)'%(density.covariance))
    plt.grid()
    return (hd, density, h)
```

# Normalization

```python
m_data = np.mean(data, axis=0)
s_data = np.std(data, axis=0)
data = (data - m_data)/s_data;
```

# Isomap mapping

## Dimensões Intrínsecas

```python
#iso = IsomapBlend()
nNeigh = int(data.shape[0] * 0.01) # 1% of the sample
#iso.SetNNeighbors(nNeigh)
#x = iso.IntrinsicDimensions(data)
x = IntrinsicDimensions(data, nNeigh)
```

## Mapeamento Final

```python
nDim = 5;
iso = manifold.Isomap(nNeigh, nDim);
iso.fit(data);
Y = iso.transform(data)
isodata = pandas.DataFrame(Y, index=data.index, columns=['D'+str(i) for i in range(nDim)])
```

```python
pandas.scatter_matrix(isodata, figsize=(16,16), diagonal='kde');
```

```python
D = metrics.pairwise_distances(Y, metric='euclidean');
X = D[np.triu_indices(D.shape[0])]
X = X[np.nonzero(X)[0]]
plt.hist(X, bins=100);
plt.yscale('log')
plt.xlabel(u'Distância (Euclidiana)')
plt.ylabel('Contagem [log]')
plt.grid()
plt.title(u'Distribuição de distâncias entre amostras');
```

## K-Means Cluster

```python
nclusters = [2,3,4,5,6,7,8]
nInit = 20
si_values = np.zeros((len(nclusters), nInit))
for i, n in enumerate(nclusters):
    for j in range(nInit):
        km = cluster.KMeans(n, n_init = 1, init='k-means++')
        km.fit(Y)
        si_values[i,j] = metrics.silhouette_score(Y, km.labels_, metric='euclidean')
        
plt.errorbar(nclusters, np.mean(si_values, axis=1), np.std(si_values, axis=1))
plt.xlim([nclusters[0]-0.5, nclusters[-1]+0.5])
plt.xlabel('# Clusters')
plt.ylabel(u'Silhueta Média')
plt.title(u'Avaliação de Clusters')
plt.grid()
```

### Number of Clusters: 3

```python
nBestClus = 3
km = cluster.KMeans(nBestClus, n_init = 20, init='k-means++')
km.fit(Y)
s = metrics.silhouette_samples(Y, km.labels_)
plot_silhouette(s, km.labels_)
plt.grid()
```

#### Clusters on Isomap Dimensions

```python
nfigs = int(sc.misc.comb(nDim,2))
Nx = 4
Ny = int(np.floor(nfigs/3.0))+1
plt.figure(figsize=(20,16))
for ifig, dims in enumerate(itertools.combinations(range(nDim),2)):
    plt.subplot(Nx,Nx, ifig+1)
    for iclus in np.unique(km.labels_):
        idx = km.labels_ == iclus
        plt.plot(Y[idx, dims[0]], Y[idx, dims[1]], '.', label='C'+str(iclus))
    plt.xlabel('D'+str(dims[0]))
    plt.ylabel('D'+str(dims[1]))
    plt.legend(numpoints=1, loc='best')
    plt.grid()
```

#### Property distributions

```python
rawdata = data * s_data + m_data
evtpred = km.predict(Y)
covfac = None # Gaussian KDE covariance factor
Nx = 5
Ny = int(rawdata.shape[1]/float(Nx)) + 1
plt.figure(figsize=(25, 20))
for icol, col in enumerate(rawdata.columns):
    plt.subplot(Ny, Nx, icol+1)
    X = np.arange(np.min(rawdata.iloc[:,icol]),
                  np.max(rawdata.iloc[:,icol]),
                  (np.max(rawdata.iloc[:,icol])-np.min(rawdata.iloc[:,icol])) * 0.01)
    for iclus in np.unique(evtpred):
        density = gaussian_kde(rawdata.iloc[np.nonzero(evtpred == iclus)[0],icol])
        if covfac:
            density.covariance_factor = lambda : covfac
            density._compute_covariance()
        plt.plot(X,density(X), '-', lw=3, label='Cl-%i'%iclus)
    
    plt.xlabel('Data Values')
    plt.ylabel('Normalized Count')
    plt.legend(loc='best')
    plt.grid()
    plt.title('Prop.: %i'%col) 
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```
